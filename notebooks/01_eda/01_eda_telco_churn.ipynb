```python
def tenure_cohort(tenure):
    if tenure <= 6:
        return "0-6 months"
    elif tenure <= 12:
        return "7-12 months"
    elif tenure <= 24:
        return "13-24 months"
    elif tenure <= 48:
        return "25-48 months"
    else:
        return "49+ months"

df["TenureCohort"] = df["TenureMonths"].apply(tenure_cohort)

```

```python
df["TenureCohort"].value_counts()
```

```python
cohort_churn = df.groupby("TenureCohort")["ChurnValue"].mean() * 100
cohort_churn
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=cohort_churn.index,
    y=cohort_churn.values,
    hue=cohort_churn.index,
    palette="viridis",
    legend=False
)
plt.xticks(rotation=45)
plt.title("Churn Rate by Tenure Cohort")
plt.ylabel("Churn %")
plt.show()

```
